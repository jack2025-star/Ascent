## 字典
- { 键:值 }
- 键名 key
- 键不重复，值可以重复

```python
students = {
    "name": "小明",
    "age": 10,
    "sex": "男",
    "score": 95,
    "is_pass": True
}

print(students["name"])
```

```python
# 访问
print(students.get("age"))
print(students.get("grade","未知")) # 不存在则返回默认值
```

```python
# 添加、修改
students["grade"] = "二年级"
students["score"] = 98
print(students)
```

```python
# 删除
del students["is_pass"]
age = students.pop("age")
print(students)
```

```python
# 遍历字典
for key in students:
    print(key,students[key])
```

```python
# 列表嵌套字典
students = [
    {"name": "小明", "age": 10},
    {"name": "小红", "age": 11},
    {"name": "小刚", "age": 12}
]

# 访问列表中字典的元素小明
print(students[0]["name"])
```

```python
#遍历
for student in students:
    print(f"姓名:{student["name"]},年龄:{student["age"]}")
```

```python
# 字典嵌套列表
reports = {
    "日期":"2026-05-31",
    "销量":[
        100,
        150,
        200
    ],
    "地区":[
        "上海",
        "北京",
        "广州"
    ]
}
print(reports["日期"])
print(reports["销量"])
```

```python
# 练习1：列表操作
prices = [10.5, 20.3, 15.8, 8.0, 25.5]
print(f"共 {len(prices)} 个商品")
print(f"最贵：{max(prices)}，最便宜：{min(prices)}")
print(f"总价：{sum(prices):.1f}，均价：{sum(prices)/len(prices):.1f}")
```

```python
# 练习2：字典操作
employee = {"name": "李华", "dept": "数据部", "salary": 15000}
employee["bonus"] = 3000
print(f"{employee['name']} 总收入：{employee['salary'] + employee['bonus']}")
```

```python
# 练习3：嵌套
orders = [
    {"product": "键盘", "qty": 2, "price": 199},
    {"product": "鼠标", "qty": 1, "price": 99},
    {"product": "显示器", "qty": 1, "price": 1299}
]
total = sum(o["qty"] * o["price"] for o in orders)
print(f"订单总金额：{total} 元")
```

```python
# 4. 列表是可变对象（注意引用）
a = [1, 2, 3]
b = a                   # b 和 a 指向同一个列表
b.append(4)
print(a)                # [1, 2, 3, 4]——a也被改了
c = a.copy()            # 用 copy 创建独立副本
```

```python
# 合并两个字典
dict1 = {"a":1,"b":2}
dict2 = {"c":3,"d":4}

dict3 = dict1 | dict2
print(dict3)

dict4 = {**dict1,**dict2}
print(dict4)
```

## 循环

```python
# 遍历列表
fruits = ["苹果", "香蕉", "橘子"]
for fruit in fruits:
    print(f"我喜欢{fruit}")
```

```python
# 遍历字符串
for char in "Python":
    print(char, end=" ")
```

```python
# 遍历字典
student = {"name": "小明", "age": 22, "city": "成都"}
for key, value in student.items():
    print(f"{key} = {value}")
```

```python
# 遍历字典的值
for value in student.values():
    print(value)
```

```python
# range(n)：0 到 n-1
for i in range(5):
    print(i, end=" ") 
```

```python
# range(start, end, step)：带步长
for i in range(0, 10, 2):
    print(i, end=" ") 
```

```python
# 倒序
for i in range(5, 0, -1):
    print(i, end=" ") 
```

```python
# enumerate：带索引遍历
names = ["小明", "小红", "小刚"]
for i, name in enumerate(names):
    print(f"{i}: {name}")
```

```python
# zip：并行遍历多个列表
names = ["小明", "小红", "小刚"]
scores = [95, 88, 72]
cities = ["成都", "北京", "上海"]

# 同时遍历
for name, score, city in zip(names, scores, cities):
    print(f"{name}来自{city}，成绩{score}")
```

```python
# 构造字典（面试常考）
name_score = dict(zip(names, scores))
print(name_score) 
```

```python
# zip 以最短的列表为准
a = [1, 2, 3, 4]
b = ["a", "b", "c"]
print(list(zip(a, b))) 
```

```python
# 当条件为 True 时持续执行
count = 0
while count < 5:
    print(count, end=" ")
    count += 1
```

```python

# break：立刻跳出整个循环
for i in range(10):
    if i == 5:
        break           
    print(i, end=" ")   



```

```python
# continue：跳过本次，继续下一次
for i in range(10):
    if i % 2 == 0:
        continue        
    print(i, end=" ")   
```

```python
# for  else：循环正常结束时执行（没被 break）
for i in range(5):
    if i == 10:
        break
else:
    print("循环正常结束，没找到 10")
```

```python
# 练习1：计算列表总和
prices = [10.5, 20.3, 15.8, 8.0, 25.5]
total = 0
for p in prices:
    total += p
print(f"总价：{total:.1f}")
```

```python
# 练习2：找出第一个大于90的成绩
scores = [78, 85, 92, 67, 95, 88]
for i, s in enumerate(scores):
    if s > 90:
        print(f"第{i+1}个学生的成绩 {s} 超过90")
        break
```

```python
# 练习3：统计词频
words = ["苹果", "香蕉", "苹果", "橘子", "香蕉", "苹果"]
counter = {}
for w in words:
    counter[w] = counter.get(w, 0) + 1
print(counter)
```

```python
# 练习4：生成成绩报告
names = ["小明", "小红", "小刚"]
scores = [95, 88, 72]
for name, score in zip(names, scores):
    grade = "A" if score >= 90 else ("B" if score >= 80 else "C")
    print(f"{name}: {score}分,等级：{grade}")
```
