# 1. str类概述

（1）pandas中提供了处理字符串操作专门的类（str）,但是这个类**只对Series对象和Index对象有效，对DataFrame对象无效。**<br>  
（2）str类中提供了50多种字符串处理方法。总的来说，这些方法和python原生语法中字符串处理方法类似。  只能作用于一维数据  
（3）注意：如果要同时使用str类下面的多个方法实现链式编程写法，则**每个方法前面均须要完整的写上str.方法名()**。如：series.str.方法1().str.方法2().str.方法3()...。  
（4）str类中的split()、replace()、contains()、count()均支持多个字符的“或”操作，但是需要**转义符（\）和或运算符（|）配合使用**，不能只写|。  
（5）类似的效果可以通过Series.map()函数实现。

# 6. 字符串的分割

| 序号 | 方法        | 作用                                                 | 示例                                                         | 主要参数                                                     | 其他说明                                                     |
| ---- | :----------- | :---------------------------------------------------- | :------------------------------------------------------------ | :------------------------------------------------------------ | :------------------------------------------------------------ |
| 1    | str.split() | 按照给定的符号**（默认是空格）**，将字符串进行分割。 | s.str.split('+')<br />s.str.split('\\+\|-\| ', expand=True) | pat：指定的分割符号（默认是空格。<br />n：分词的次数。None、0、-1表示分割整个字符串<br />expand：是否将分割后的字符串放在不同的列（扩展为一个DataFrame对象）。 | pat支持用**“\|”**符号连接多种不同的字符进行分割。但是，**必须在最前面加转义符**。 |

```python
s = pd.Series(['1-2-3-4-5-6', 'A+B+C+D+E+F', 'a b c d e f'])
s
```

```python
# 按照+分割
s.str.split('+')
```

```python
# 按照+分割2次  ,前两个
s.str.split('+',n = 2)
```

```python
# 按照-分割并将分割后的字符串放在不同的列    Dateframe
s.str.split('-',expand = True)
```

```python
# 重要转义符：\ ，同时 regex: 'bool | None' = None，可以表示为正则表达式分割
# 按照+ - 空格分割并将分割后的字符串放在不同的列
s.str.split(r'\+|-| ',expand = True)
```

# 7. 字符串的替换

str类下面的str.replace()方法与pandas中Series对象和DataFrame对象的replace()方法不同，前者是替换字符串里面的符号，而后者是替换字符串。前者类似于Excel表中的普通替换模式，后者类似于Excel表中的单元格匹配替换模式。

| 序号 | 方法          | 作用                             | 示例                          | 主要参数                                                     | 其他说明                                                     |
|:---- | :------------- | :-------------------------------- | :----------------------------- | :------------------------------------------------------------ | :------------------------------------------------------------ |
| 1    | str.replace() | 将字符串中的符号替换成其他符号。 | s.str.replace('\+\|-\| ','=') | pat：被替换的字符。<br />repl：替换后的新字符。<br />case：是否大小写敏感，默认区分。<br /> | pat支持用“\|”符号连接多种不同的字符进行分割。但是，**必须在最前面加转义符**。 |

```python
df = pd.DataFrame(
    [['ABC', 'ABD', 'ABC'], ['ABE', 'FGH', 'ABD']], columns=['第一列', '第二列', '第三列'])
df
```

```python
# DataFrame对象的replace()方法，支持inplace参数
# 将'ABC'转换为'ABK'
df.replace('ABC','ABK')
```

```python
# 将'AB'转换为'12'
```

```python
s = pd.Series(['1-2-3-4-5-6', 'A+B+C+D+E+F', 'a b c d e f'])
s
```

```python
# 将+替换为-
s.str.replace('+','-')
```

```python
# 将+ - 和空格都替换为=，其中regex: 'bool' = False，此时需要显示指定regex= True
s.str.replace('\+|-|','=',regex = True)
```

# 8. 字符串的匹配

| :---- | :---------------- | :-------------------------- | :------------------------------ | :--------------------------------------------- | :---------------- |
| 1    | str.contains()   | 判断是否包含某个字符串。   | s.str.contains('A',case=False) | pat：匹配字符串。<br />case：默认区分大小写。 |                  |

```python
s = pd.Series(['ABC','abc','BcD','bcd'])
```

```python
# 是否包含B或D，不区分大小写
s.str.contains('B|D', case=False)
```

```python
# 是否以Bc开头
s.str.startswith('Bc')
```

```python
# 是否以D结尾
s.str.endswith('D')
```

# l0.练习

```python
data = {
    '姓名': ['孙强', '王小明', np.nan, '李华', '朱琳', '赵丽', np.nan, '张伟', '陈静', '吴刚'],
    '性别': ['男', '男', '男', 'f', '男', 'm', 'm', 'm', '男', '女'],
    '语文成绩': ['77', '74', np.nan, '95', '97', '75', '91', '75', '94', '67'],
    '数学成绩': ['74', '98', '82', '100', '75', '74', np.nan, '95', '81', '99'],
    '身份证号': [
        '110101-19970219-5320', 
        '863794026542351234', 
        '330102 880124 570802', 
        '931034131647525534', 
        np.nan, 
        '440301920503316932', 
        np.nan, 
        '965328710122691782', 
        '440301930315103420', 
        '440301-850925-643020'
    ]}
df = pd.DataFrame(data)
df
```

（1）将姓名列缺失值填充为'未知'  
（2）统一性别格式，将f转换为女，m转换为男  
（3）将语文成绩和数据成绩转换为数值型数据，同时对于缺失部分填充为0  
（4）清洗身份证号  
（5）计算男、女语文数学平均成绩

```python
# （1）将姓名列缺失值填充为'未知'
df['姓名'] = df['姓名'].fillna('未知')
df
```

```python
# （2）统一性别格式，将f转换为女，m转换为男
# 方法1
df['性别'].replace({'f':'女','m':'男'},inplace = True)
df
```

```python
# 方法2
df['性别'] = df['性别'].str.replace('m','男').str.replace('f','女')
df
```

```python
# 数据集基本信息
df.info()
```

```python
# （3）将语文成绩和数据成绩转换为数值型数据，同时对于缺失部分填充为0
df['语文成绩'] = pd.to_numeric(df['语文成绩'],errors = 'coerce').fillna(0)
df['数学成绩'] = pd.to_numeric(df['数学成绩'],errors = 'coerce').fillna(0)
df
```

```python
# （4）清洗身份证号
df['身份证号_清洗后'] = df['身份证号'].str.replace(r'[^0-9]','',regex = True)
df
```

```python
# （5）计算男、女语文数据平均成绩
df.groupby(by = '性别').agg({'语文成绩':'mean','数学成绩':'mean'})
```
