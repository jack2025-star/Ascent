# 3.数据的引用

## 3.1数据的引用概述

### 3.1.1索引的内涵

（1）所谓索引（名词），就是与值对应的**“标签”**，标签可以是：**程序自动生成的、数据对象自带的、默认的位置序号（从0开始，正向递增的整数；或从-1开始反向递减的整数）**；标签也可以是用**户自定义的符号，数字、字符串等均可**。并且，在pandas中，这**两种方式同时存在，同时有效**。<br>
（2）DataFrame对象是二维表结构，既有**行索引**，也有**列索引**。口语中，只说索引，则表示行索引；而称列索引为列名。

### 3.1.2数据引用的内涵

（1）数据的引用是指从完整的DataFrame对象中**选择或筛选**出特定的数据，如选择某一列或某一行，或者按照条件筛选某几列或某几行。   
（2）Series对象和DataFrame对象数据的引用符号主要包括**三种符号**：<br>
1）**普通的[ ]符号**。如df[列名]或df[[列1，列2...]]可用于取出某一列或某几列（**注意：此方法只能用在列索引操作或行切片操作（不能用在列切片操作或行索引操作）**）。  
2）**.iloc[ ]符号**。按照**程序自带的位置索引**（integer index）进行引用。  
3）**.loc[ ]符号**。按照**用户指定的行索引或列索引（列名）**进行引用（**注意：iloc[]或loc[]的[]中，标准的写法应该是[行索引/切片,列索引或切片]**），当对行进行操作时，如果不筛选列，则 ：符号可以省略不写；反之则“不行”**）。    
（3）普通的[ ]符号以及iloc[ ]符号的[ ]中还可以传入**一组Boolean类型的值，实现条件引用**，筛选出符合条件的数据。  
（4）通过与where()、mask()、query()等方法的组合，可实现更多引用功能。

## 3.2几种常见的操作

在实际操作中，对于数据的引用一般有如下几类操作需求：  
（1）根据行、列数据是否被完全筛选，可以分为：引用行（不筛选列）、引用列（不筛选行）、既引用行又引用列。  
（2）根据行、列引用数量的多少，可以分为：  
1）索引（动词）：选择某一行或某一列。  
2）切片：选择连续的某几行或连续的某几列，或根据步长有规律选择不连续的某几行或某几列。  
3）选择指定的某几行或某几列。  
4）选择满足条件的行或列（条件筛选）。
