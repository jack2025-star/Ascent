```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'SimHei'    # 设置中文显示
plt.rcParams['axes.unicode_minus']=False  # 该语句解决图像中的“-”负号的乱码问题
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

此处所说的多图绘制包括三种情况：  
（1）同时在一段代码中创建多个画布（figure），绘制多个图形。  
（2）在一个画布（figure）中创建多个子图（subplots）。  
（3）在一个子图（subplot）中绘制多个图例（legend）。

# 1. 多图形绘制

核心方法就是创建多个plt.figure()，在每个plt.figure()下面独立绘制图形即可。

```python
# 创建两个画布，画布1，大小为（9,3），绘制y=x的折线图
plt.figure(num = 1 , figsize = (9,3))
plt.plot(np.arange(10))
# plt.show()

# 画布2，大小为（9,3），绘制y=x^2的折线图
plt.figure(num = 2, figsize = (9,3))
plt.plot(np.arange(10)**2)
```

# 2. 多子图绘制

核心方法是使用plt.subplot()或plt.subplot2grid()创建多个subplot，然后在每个plt.subplot()或plt.subplot2grid()下面独立绘制图形即可。注意：plt.subplot()中子图的序号从1开始，而plt.subplot2grid()中子图的xy坐标从0开始。

```python
# 创建一个画布，一行三列三个子图，画布大小（9,3）
plt.figure(figsize = (9,3))
# 子图1，绘制y=x的折线图
plt.subplot(1,3,1)
plt.plot(np.arange(10))
# 子图2，绘制y=x^2的折线图
plt.subplot(1,3,2)
plt.plot(np.arange(10)**2)
# 子图3，绘制y=x^3的折线图
plt.subplot(1,3,3)
plt.plot(np.arange(10)**3)
plt.show()
```

```python
# 创建一个画布，一行2行2列子图，画布大小（9,3）
plt.figure(figsize = (9,3))
# 子图1，起始坐标（0,0），横跨2列，绘制y=x的折线图
plt.subplot2grid((2,2),(0,0),colspan = 2)
plt.plot(np.arange(1,10))

# 子图2，起始坐标（1,0），绘制y=x^2的折线图
plt.subplot2grid((2,2),(1,0))
plt.plot(np.arange(1,10)**2)

# 子图3，起始坐标（1,1），绘制y=x^3的折线图
plt.subplot2grid((2,2),(1,1))
plt.plot(np.arange(1,10)**3)

# 子图直接宽度和高度间隔0.5
plt.subplots_adjust(wspace = 0.5,hspace = 0.5)
plt.show()
```

# 3. 多图例绘制

核心是在一个子图下绘制多个图例。

```python
# 创建一个画布，绘制2个子图，在一幅图里绘制2条折线图，画布大小（9,3）
plt.figure(figsize = (9,3))

# 子图1，绘制绘制y=x和y=x^2的折线图

plt.subplot(1,2,1)               # 2. 选择左边区域 (1行2列布局中的第1个)
plt.plot(np.arange(1,10))
plt.plot(np.arange(1,10) ** 2)

# 子图2，绘制绘制y=x^3和y=x^4的折线图

plt.subplot(1,2,2)            # 3. 选择右边区域 (1行2列布局中的第2个)          
plt.plot(np.arange(1,10) ** 3)
plt.plot(np.arange(1,10) ** 4)
plt.show()
```

# 4.双Y轴图

```python
# 第一步：创建图形对象
fig = plt.figure(figsize = (10,6))

# 第二步：添加子图
ax1 = fig.add_subplot(1,1,1)
      

# 数据
months = ['1月', '2月', '3月', '4月', '5月', '6月']
sales = [100, 150, 120, 180, 220, 250]
growth = [5, 8, 3, 12, 15, 10]

# 绘制左侧Y轴折线图，月份和销售额，蓝色实线实心圆，宽度2，标签'销售额(万元)'
ax1.plot(months,sales,'b-o',lw = 2, label = '销售额(万元)')

# y轴标签'销售额(万元)'颜色蓝色
ax1.set_ylabel('销售额(万元)',color = 'b')
# Y轴刻度标签的颜色为蓝色
ax1.tick_params(axis = 'y',labelcolor = 'b')

# 绘制右侧Y轴折线图，月份和增长率，红色正方形，宽度2，标签'增长率(%)'
# 使用同一个x轴
ax2 = ax1.twinx()
ax2.plot(months,growth,'r--s',lw = 2,label = '增长率（%）')


# y轴标签'增长率(%)'颜色红色
ax2.set_ylabel('增长率%',color = 'r')

# Y轴刻度标签的颜色为红色
ax2.tick_params(axis = 'y', labelcolor = 'r')

# 标题'销售额与增长率双Y轴图'
plt.title('销售额与增长率双Y轴图')
    
# 网格线
plt.grid(True,alpha = 0.3)

#图例
ax1.legend(loc = 'upper left',bbox_to_anchor = (0,1))
ax2.legend(loc = 'upper left',bbox_to_anchor = (0,0.9))

# 自动调整子图参数，使图形元素（标签、标题、刻度等）不会重叠
plt.tight_layout()

plt.show()
```
