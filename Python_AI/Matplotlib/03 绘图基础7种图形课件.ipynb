# plt.bar()函数基础

plt.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

    x：表示x坐标，数据类型为int或float类型，刻度自适应调整；也可传dataframe的object，x轴上等间距排列；
    height：表示柱状图的高度，也就是y坐标值，数据类型为int或float类型；
    width：表示柱状图的宽度，取值在0~1之间，默认为0.8；
    bottom：柱状图的起始位置，也就是y轴的起始坐标；
    align：柱状图的中心位置，默认"center"居中，可设置为"edge"边缘；
    color：柱状图颜色；
    edgecolor：边框颜色；
    linewidth：边框宽度；
    alpha:透明度，值越小越透明;
    tick_label:指定条形图的刻度标签 
    label:指定条形图的标签，一般用以添加图例

```python
# 绘制四个直辖市GDP条形图，画布大小（9,3）,GDP=[12406.8,13908.57,9386.87,9143.64],x=range(4)
# 柱状图的中心位置为中心，颜色为蓝色，透明度为0.8
# y轴标签GDP，图片标题四个直辖市GDP大比拼，x轴刻度ticks=range(4),labels=['北京市','上海市','天津市','重庆市']，y轴范围[5000,15000]。
plt.figure(figsize=(9,3))
# 构建数据
GDP = [12406.8,13908.57,9386.87,9143.64]
x = range(4)

# 绘图
plt.bar(x = x,height = GDP,align = 'center',color = 'b',alpha = 0.8)

# 添加轴标签
plt.ylabel('GDP')

# 添加标题
plt.title('四个直辖市GDP大比拼')

# 添加刻度标签
plt.xticks(ticks = x, labels = ['北京市','上海市','天津市','重庆市'])

# 设置Y轴的刻度范围
plt.ylim(5000,15000)

# 为每个条形图添加数值标签
for i,j in zip(x,GDP):
    plt.text(i,j + 100,str(j),ha = 'center')
    
# 显示图形
plt.show()
```

# plt.barh()函数基础

plt.barh(
    y,
    width,
    height=0.8,
    left=None,
    *,
    align='center',
    data=None,
    **kwargs,
)

y: 条形图的y坐标，即条形的垂直位置。  
width: 条形的宽度（水平方向的长度）。  
height: 条形的高度（垂直方向的长度），默认0.8。  
left: 条形的左侧起始位置，即条形的起始x坐标。默认None，表示从0开始。  
align: 条形的对齐方式，默认'center'，即条形的中心位于y值处。也可以设为'edge'，即条形的边缘位于y值处。  
data: 用于指定数据来源，如果提供了，则可以直接用字符串指定数据框的列名。  

```python
# 绘制水平条形图，数据为price = [39.5,39.9,45.4,38.9,33.34]
# 柱状图的中心位置为中心，颜色为steelblue，透明度为0.8
# x轴标签价格，图片标题不同平台书的最低价比较，y轴刻度ticks=range(4),labels=['亚马逊','当当网','中国图书网','京东','天猫']，x轴范围[32,47]。
# 构建数据
price = [39.5,39.9,45.4,38.9,33.34]

# 绘图
plt.barh(y = range(5),width = price,align = 'center',color = 'steelblue',alpha = 0.8)

# 添加轴标签
plt.xlabel('价格')

# 添加标题
plt.title('不同平台书的最低价比较')

# 添加刻度标签
plt.yticks(ticks=range(5),labels=['亚马逊','当当网','中国图书网','京东','天猫'])

# 设置X轴的刻度范围
plt.xlim(32,47)

# 为每个条形图添加数值标签
for i,j in enumerate(price):
    plt.text(j+0.1,i,str(j),va = 'center')

# 显示图形    
plt.show()
```

# plt.hist()函数基础

plt.hist(x, bins=10, range=None, normed=False,weights=None, cumulative=False, bottom=None,histtype='bar', align='mid', orientation='vertical',rwidth=None, log=False, color=None,label=None, stacked=False)

    x：指定要绘制直方图的数据；
    bins：指定直方图条形的个数；
    range：指定直方图数据的上下界，默认包含绘图数据的最大值和最小值；
    normed：是否将直方图的频数转换成频率；
    weights：该参数可为每一个数据点设置权重；
    cumulative：是否需要计算累计频数或频率；
    bottom：可以为直方图的每个条形添加基准线，默认为0；
    histtype：指定直方图的类型，默认为bar，除此还有’barstacked’, ‘step’, ‘stepfilled’；
    align：设置条形边界值的对其方式，默认为mid，除此还有’left’和’right’；
    orientation：设置直方图的摆放方向，默认为垂直方向；
    rwidth：设置直方图条形宽度的百分比；
    log：是否需要对绘图数据进行log变换；
    color：设置直方图的填充色；
    label：设置直方图的标签，可通过legend展示其图例；
    stacked：当有多个数据时，是否需要将直方图呈堆叠摆放，默认水平摆放；

# plt.boxplot()函数基础

    boxprops：设置箱体的属性，如边框色，填充色等；
    labels：为箱线图添加标签，类似于图例的作用；
    filerprops：设置异常值的属性，如异常点的形状、大小、填充色等；
    medianprops：设置中位数的属性，如线的类型、粗细等；

# plt.scatter()函数基础

plt.scatter(
    x,
    y,
    s=None,
    c=None,
    marker=None,
    cmap=None,
    norm=None,
    vmin=None,
    vmax=None,
    alpha=None,
    linewidths=None,
    *,
    edgecolors=None,
    plotnonfinite=False,
    data=None,
    **kwargs,
)

x, y: 数组，表示散点的横纵坐标。  
s: 标量或数组，表示散点的大小。如果是标量，则所有点大小相同；如果是数组，则对每个点设置不同大小。  
c: 颜色。可以是颜色字符串、颜色序列，或者一系列数值，然后通过cmap映射。如果是数值数组，则会根据norm和cmap进行颜色映射。  
marker: 散点形状。    
cmap: 颜色映射。当c是数值数组时，用于指定颜色映射。  
norm: 归一化方法。用于将c的数值映射到颜色映射的范围。通常使用Normalize类或其子类。  
vmin, vmax: 标量。与norm结合使用，用于设置颜色映射的范围。如果norm被指定，则忽略vmin和vmax。  
alpha: 透明度，介于0（透明）和1（不透明）之间。  
linewidths: 标量或数组，表示散点边缘的线宽。  
edgecolors: 边缘颜色。可以是颜色字符串、颜色序列或'face'（与内部颜色相同）、'none'（无边缘）。  

```python
# 生成销售数据
np.random.seed(123)   #随机数的种子，随机数字是可重复
months = np.arange(1, 13)
sales_A = np.random.normal(100, 20, 12)  # 产品A销售额
profit_margin = np.random.uniform(0.1, 0.3, 12)  # 利润率
# 创建画布（10,6）
plt.figure(figsize=(10, 6))

# 创建散点图：x=月份, y=销售额, 大小=销售额, 颜色=利润率
sc = plt.scatter(x = months,y = sales_A,s = sales_A,c = profit_margin,cmap = 'RdYlGn',alpha = 0.7,
                 edgecolors = 'black',linewidth = 0.5)


                               # 点大小表示销售额
                          # 颜色表示利润率
                           # 红-黄-绿色系
                                 
# 添加颜色条
char = plt.colorbar(sc,label = '利润率')

# 美化图表
plt.xlabel('月份')
plt.ylabel('销售额（万元）')
plt.title('产品A月度销售额分析')
plt.grid(True,alpha =  0.3)
plt.tight_layout()
plt.show()
```
