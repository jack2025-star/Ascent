# 2. matplotlib的核心逻辑

## 2.1 figure、subplot/Axes、axis的理解

要理解matplotlib的核心逻辑，掌握matplotlib的绘图步骤，就必须理解：figure、subplot/Axes、axis等概念。  
（1）**figure：画布**。画布可以理解为一张空白的画纸；想要绘图，则首先**必须**要设置一个画布。不过，如果只需要一个画布绘制图形，则可以不用显式定义plt.figure()，系统会**默认**创建一个figure。但是，如果要在不同的画布上绘制图形，则需要显式定义plt.figure()。  
（2）**subplot/Axes：子图、轴域、坐标系等。**subplot和Axes的概念非常类似，即画纸上的绘图区域。设置完figure后，**必须**设置subplot/Axes。不过，如果在一个figure上只绘制一张图，则可以不用显式设置subplot/Axes，系统会**默认**创建一个subplot/Axes。但是，如果要在同一个figure上绘制多个图形，则必须显式添加subplot/Axes。  
（3）**axis：轴。**Axes是axis的复数形式，axis通常只特定的某个轴，如x轴，y轴等等，由多个axis就构成了一个Axes。

**【再次强调】必须要定义figure，设置（添加）subplot/Axes**。只不过，如果只是在一个画布（figure）上绘制一幅图形（subplot/Axes），则不需要显式定义（系统隐式定义），系统会**默认**自动生成。而当需要在多个画布（figure）上绘画，或一个画布（figure）上绘制多个图形（subplot/Axes），则必须显式设置（添加）subplot/Axes。

## 2.2 一幅典型图形的组成元素

```python
# 导入相关的库
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
%matplotlib inline
 #图像能显示中文
 #确保负坐标轴能显示负号


# 准备数据
x=[-23,-13,-4,6,10,18,28,35,40]
y=[12,15,16,17,17,18,25,28,38]

# 绘制图形
# 设置画布
#figsize画布大小，dpi分辨率，faceclor画布颜色 
plt.figure(figsize=(10,5),dpi = 100,facecolor = 'white')
# 添加图形
plt.plot(x,y,
         lw = 2,
         ls = '--',
         c = 'r',
         marker = 'o',
         markerfacecolor = 'blue',
         label='曲线')
         

        #线宽度
         #线型
         #线的颜色
        #数据点形式
         #数据点大小
         #数据点颜色
         
#添加标题
plt.title('x-y图')

#添加x轴说明
plt.xlabel('x轴')

#添加y轴说明
plt.ylabel('y轴',rotation = 0)

#x轴的范围
plt.xlim(-25,45)

#y轴的范围
plt.ylim(10,40)

# X轴的间隔，rotation旋转角度
plt.xticks(ticks = [-30,-10,10,30,50])

# Y轴的间隔
plt.yticks(ticks = [10,20,30,40])

#添加网格
plt.grid(True,linestyle = '--',alpha = 0.3)

#添加图例，loc--可以指定图例的位置,'best'自动选择最优
plt.legend(loc = 0)

# 显示图形
plt.show()
```

```python
# 多图绘制
# 设置画布
plt.figure(num = 1,figsize = (12,4))

# 子图1（subplot1）
  # 注意：子图顺序从1开始。
plt.subplot(1,2,1)
# 绘制y=x的折线图，标记为实心圆
plt.plot(range(10),marker = 'o')

# 标题'y=x'，x轴标签x，y轴标签y，字体大小12
plt.title('y == x')
plt.xlabel('x',fontsize = 12)
plt.ylabel('y',fontsize = 12)


# 子图2（subplot2）
plt.subplot(1,2,2)

# 绘制y=x^2与y=x^3的折线图，宽度为3，y=x^2为破折线红色，标签为'$y=x^2$'，y=x^3为点划线绿色，标签为'$y=x^3$'
plt.plot(np.arange(10)**2,lw = 3,ls = '--',c = 'r',label = '$y=x^2$')
plt.plot(np.arange(10)**3,lw = 3,ls = '-.',c = 'g',label = '$y=x^3$')

# 标题'$y=x^2$与$y=x^3$'，x轴标签x，y轴标签y，字体大小12
plt.title('$y=x^2与$y=x^3$')
plt.xlabel('x',fontsize = 12)
plt.ylabel('y',fontsize = 12)
# 显示图例
plt.legend()

# 保存
# dots per inch图片保存大小问题，设置bbox_inches=‘tight’
plt.savefig('./image/图片.png',dpi = 300,bbox_inches = 'tight')
# 特别注意：plt.show()一定要放在plt.savefig()后面
plt.show()
```

```python
# 创建画布，画布大小（20,5）
plt.figure(figsize=(20,5))  
# 绘制1行4列四个子图
plt.subplot(1,4,1)
plt.subplot(1,4,2)
plt.subplot(1,4,3)
plt.subplot(1,4,4)

# 每个子图之间宽度0.5
plt.subplots_adjust(wspace = 0.5)
```

```python
# 创建3行4列的子图格子，第一个子图起始位置为（0，0）横跨4列占一行
plt.subplot2grid((3,4),(0,0),colspan = 4)

# 第二个子图起始位置为（1，0）竖直跨两行，占一列
plt.subplot2grid((3,4),(1,0),rowspan = 2)

# 第三个子图起始位置为（1，1）竖直跨两行，横跨3列
plt.subplot2grid((3,4),(1,1),rowspan = 2,colspan = 3)

# 子图直接宽度间隔0.5，高度间隔0.5
plt.subplots_adjust(wspace = 0.5,hspace = 0.5)
```

```python
# 绘制1-12月销售额的两条折线图
x = range(1, 13)
y1 = np.random.randint(1, 10, 12)
y2 = np.random.randint(5, 10, 12)

# 创建图形
plt.figure(figsize=(12,6))

# 折线图1，标签为A
plt.plot(x,y1,label = 'A')
# 折线图2，标签为B
plt.plot(x,y2,label = 'B')

# 增加X轴刻度标签
plt.xticks(ticks=range(1,13))
plt.title('1-12月销售额',fontdict = {'family':'SimHei','fontsize':15},loc = 'center',pad= 15)
plt.xlabel("月份")
plt.xlabel("销售额（万元）")

# 添加图例
plt.legend(loc = 0,prop = {'family':'SimHei','size':12})

# 为两条折线添加数据标签，调整位置避免重叠
# (x,y1+0.2)给y1加标签，黑体，10号，加粗，水平居中，垂直底部对齐
for i,j in zip(x,y1):
    plt.text(i,j+0.2,str(j),
           fontdict={'family':'SimHei','fontsize':10,'weight':'bold'},
           ha='center',va='bottom')
             

# (x,y2-0.3)给y2加标签，黑体，10号，加粗，水平居中，垂直底部对齐
for a,b in zip(x,y2):
    plt.text(a,b-0.3,str(b),fontdict={'family':'SimHei','fontsize':10,'weight':'bold'},
           ha='center',va='bottom')


    
plt.grid(True,alpha = 0.3,linestyle = '--')
# 添加网格线提高可读性
plt.show()
```

## 4.1 线条的格式控制

设置pyplot的动态rc参数

线条的格式控制一般包括**“颜色”、“风格”、“宽度”、“标记”**等。  

（1）颜色：使用**color参数（可简写为c）**控制。参数的设置可以有多种方式，常见的包括：  
1）颜色的**英文单词**，如red, green, black等等。  
2）颜色的**英文单词的简写**，如r,g,b,c等等。  
3）颜色的**RGB值（0-1区间的十进制数字）**：如(0,0.1,0.2)，(0.2,0.3,0.5)等等。**注意：必须是0-1区间上的数字**。<br>
4）颜色的**RGB值（十六进制的字符串）**：如'#000000'，'#FF0000'，'#FF00FF'等等。**注意：字符串前面必须写#号**。<br>
颜色样式可参考：https://www.runoob.com/html/html-colorvalues.html

| 符号 | 'r'  | 'g'  | 'b'  | 'c'  | 'm'  | 'y'  | 'k'  | 'w'  |
| ---- | ---- | ---- | ---- | ---- | ---- | ---- | ---- | ---- |
| 颜色 | 红色 | 绿色 | 蓝色 | 青色 | 品红 | 黄色 | 黑色 | 白色 |

（2）风格：使用**linestyle参数（可简写为ls）**控制。参数的常用值如下表：

```python
# 绘制1-12月销售额的两条折线图
x = range(1, 13)
y1 = np.random.randint(1, 10, 12)
y2 = np.random.randint(5, 10, 12)

# 创建图形
plt.figure(figsize=(12, 6))

# 折线图1，标签为A,数据标记点为实心圆，线条宽度为2
plt.plot(x,y1,label = 'A')
# 折线图2，标签为B,数据标记点为像素点，线条宽度为2
plt.plot(x,y2,label = 'B')

# 增加X轴刻度标签为1-12
plt.xticks(ticks=range(1, 13))
plt.title('1-12月销售额', fontdict={'family':'SimHei','fontsize':15}, loc='center', pad=15)
plt.xlabel('月份')
plt.ylabel('销售额（万元）')

# 添加图例
plt.legend(loc='best', prop={'family': 'SimHei', 'size': 12})

# 为两条折线添加数据标签，调整位置避免重叠
# (x,y1+0.2)给y1加标签，黑体，10号，加粗，水平居中，垂直底部对齐
# 为文本添加一个背景框，'round': 圆角矩形，pad=0.2: 内边距，设置背景框的填充颜色淡蓝色，透明度0.7
for i,j in zip(x,y1):
    plt.text(i,j+0.2,str(j),
           fontdict={'family':'SimHei','fontsize':10,'weight':'bold'},
           ha='center',va='bottom')
    
# (x,y2-0.3)给y2加标签，黑体，10号，加粗，水平居中，垂直底部对齐
# 为文本添加一个背景框，'round': 圆角矩形，pad=0.2: 内边距，设置背景框的填充颜色淡珊瑚红色，透明度0.7
for a,b in zip(x,y2):
    plt.text(a,b-0.3,str(b),
           fontdict={'family':'SimHei','fontsize':10,'weight':'bold'},
           ha='center',va='bottom')

# 添加网格线提高可读性
plt.grid(True, alpha=0.3, linestyle='--')
# 调整Y轴范围，为标签留出空间
plt.savefig('./image/1-12月销售额.png',dpi=400,bbox_inches='tight')
plt.show()
```

# 6.练习

需求描述：  
（1）读取数据源中消费数据，赋值给变量df。对数据集进行初步探索。  
（2）将餐券已经使用的行筛选出来，重置索引，赋值给新的变量dfN，对dfN进行缺失值和重复值检查。    
（3）统计分析用户就餐频次前20的数据，并绘制水平条形图。  
（4）统计分析平均每天就餐时间分布，绘制折线图。   

```python
# 读取消费数据，命名为df，预览前两行
import os
print(os.getcwd()) 
df = pd.read_excel("./数据源/消费数据.xlsx")
df.head(2)
```

```python
# 查看数据集基本信息
df.info()
```

```python
# 筛选出状态等于已使用的数据并重置索引，并赋值给变量dfN
dfN = df[df['状态'] == '已使用'].reset_index(drop = True)
dfN
```

```python
# 查看是否存在缺失值
dfN.isna().sum()
```

```python
# 查看是否存在重复值
dfN.duplicated().sum()
```

用户就餐频次

```python
# 统计用户就餐频次，转化成DataFrame数据类型，取出频次前20的数据赋值给变量dfFreq。
dfFreq = dfN['用户ID'].value_counts().to_frame().nlargest(20,'count')
dfFreq
```

```python
# 将dfFreq的索引的数据类型转换成str
dfFreq.index = dfFreq.index.astype('str')
```

```python
# 绘制水平条形图
# 创建画布大小（20，20），分辨率300
plt.figure(figsize = (20,20),dpi = 300)
# 水平条形图，颜色为绿色
plt.barh(y = dfFreq.index,width = dfFreq['count'],color = 'g')
# X轴标签用户频次，大小15，Y轴标签用户ID，大小15
plt.xlabel('用户频次',fontsize = 15)
plt.ylabel('用户ID',fontsize = 15)
# 图标题出现频次最高的20位用户，大小20
plt.title('出现频次最高的20位用户',fontsize = 20)
# 给出数据标签，水平方向增加2个距离，水平居中，垂直底部，大小10
for i,j in zip(dfFreq['count'],dfFreq.index):
    plt.text(i+ 2,j,str(i),ha= 'center',va= 'bottom',fontsize= 10)
# 显示图形
plt.show()
```

平均每天就餐时间分布

```python
# 将dfN核销时间中的消费的时和分取出来，并升序排列，重置索引，赋值给变量df_useTime
df_useTime = dfN['核销时间'].str[11:16].sort_values().reset_index(drop= True)
df_useTime
```

```python
# 统计每个时间出现的频次，并按索引升序排列，赋值给变量df_useTimeFreq
df_useTimeFreq = df_useTime.value_counts().sort_index()
df_useTimeFreq
```

```python
# 绘制折线图
# 设置画布大小（10,5），分辨率300
plt.figure(figsize = (10,5),dpi = 300)
# 折线图，x轴为核销时间，Y轴为平均每天就餐人数，红色，实线
plt.plot(df_useTimeFreq.index,
         df_useTimeFreq.values/(dfN['餐卷所属日期'].nunique()),c = 'r',ls='-')
# x轴刻度取值（0,125,20）
plt.xticks(range(0,125,20))
# y轴标签每天平均就餐人数，x轴标签时间
plt.ylabel('每天平均就餐人数')
plt.xlabel('时间')
# 标题每天平均就餐时间分布，大小15
plt.title('每天平均就餐时间分布',fontsize = 15)
# 显示图像
plt.show()
```
