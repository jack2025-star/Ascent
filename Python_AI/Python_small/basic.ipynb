```python
#简易计算机
import math

try:
    num1 = float(input("请输入第一个数字: "))
    num2 = float(input("请输入第二个数字: "))
except ValueError:
    print("请输入有效的数字")
    exit()

print("请输入运算符:")
print("1加法 +")
print("2减法 -")
print("3乘法 *")
print("4除法 /")
print("5平方根 sqrt")
print("6幂运算 x^y")

choice = input("请输入选项1-6: ")

print()


if choice == '1':
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")

elif choice == '2':
    result = num1 - num2
    print(f"{num1} - {num2} = {result}")

elif choice == '3':
    result = num1 * num2
    print(f"{num1} × {num2} = {result}")

elif choice == '4':
    # 除数不能为 0
    if num2 == 0:
        print("错误：除数不能为0")
    else:
        result = num1 / num2
        print(f"{num1} ÷ {num2} = {result}")
elif choice == '5':
    try:
        result1 = math.sqrt(num1)
        result2 = math.sqrt(num2)
        print(f"结果：{num1} = {result1}, {num2} = {result2}")
    except ValueError:
        print("不能对负数求实数平方根")

elif choice == '6':
     result = num1 ** num2
     print(f"{num1} ^ {num2} = {result}")

else:
    print("错误，请重新输入")
```

## 简易通讯录
1. 目标
- 练习字典的使用，字典嵌套字典，数据的查询和修改

2. 实现
- 存储练习人信息 {姓名:{电话,邮箱,地址}}

```python
#简易通讯录
print("简易通讯录")


#初始化通讯录
contacts  = {
    "小明":{
        "phone":"13800138001",
        "email":"xiaoming@email.com",
        "address":"成都"
    },
    "小红":{
        "phone":"13800138002",
        "email":"xiaohong@email.com",
        "address":"深圳"  
    }

}

# 函数显示联系人
def show_contacts():
    print('-'*50)
    #判断字典长度
    if len(contacts) == 0:
        print("联系人为空")
        return
        # 返回None
    
    # 使用for循环遍历,字典名.items()用于获取字典的键值对
    # name 是键    info 是值
    for name,info in contacts.items():
        print(f"姓名:{name}\n")
        print(f"电话:{info['phone']}  ")
        print(f"邮箱:{info['email']}  ")
        print(f"地址:{info['address']}  ")
    
    print(f"共{len(contacts)}个联系人\n")  
    
# 添加联系人
def add_contact():
    print("添加联系人\n")
    name = input("姓名:")
    # 判断是否存在
    if name in contacts:
        print("联系人已存在")
        return
        
    phone = input("电话:")
    email = input("邮箱:")
    address = input("地址:")
    
    contacts[name] = {
        "phone":phone,
        "email":email,
        "address":address
    }
    print("已添加联系人\n")

#查找联系人
def search_contact():
    print("查询联系人\n")
    name = input("请输入需要查询的联系人：")
    
    if name in contacts:
        info = contacts[name]
        print(f"找到联系人：{name}\n")
        print(f"电话：{info['phone']}  ")
        print(f"邮箱：{info['email']}  ")
        print(f"地址：{info['address']}  ")
    else:
        print(f"未找到联系人：{name}\n")

# 删除联系人
def delete_contact():

    print("删除联系人\n")
    name = input("请输入要删除的姓名：")

    if name in contacts:
        del contacts[name]  # del删除字典中的键
        print(f"已删除联系人：{name}")
    else:
        print(f"未找到联系人：{name}")

#更新联系人
def update_contact():
    print("修改联系人\n")
    name = input("请输入要修改的姓名：")

    if name in contacts:
        print(f"当前信息：\n")
        print(f"电话：{contacts[name]['phone']} ")
        print(f"邮箱：{contacts[name]['email']} ")
        print(f"地址：{contacts[name]['address']} ")

        print("\n请输入新信息,直接回车表示保持原值：")

        phone = input("新电话：")
        if phone:
            contacts[name]['phone'] = phone

        email = input("新邮箱：")
        if email:
            contacts[name]['email'] = email

        address = input("新地址：")
        if address:
            contacts[name]['address'] = address

        print(f"已更新联系人：{name}")
    else:
        print(f"未找到联系人：{name}")


while True:
    print("操作菜单：")
    print("1. 显示所有联系人")
    print("2. 添加联系人")
    print("3. 查询联系人")
    print("4. 删除联系人")
    print("5. 修改联系人")
    print("6. 退出")
    print("-" * 50)

    choice = input("请选择操作(1-6): ").strip()

    if choice == '1':
        show_contacts()

    elif choice == '2':
        add_contact()

    elif choice == '3':
        search_contact()

    elif choice == '4':
        delete_contact()

    elif choice == '5':
        update_contact()

    elif choice == '6':
        print("\n感谢使用，再见！")
        break

    else:
        print("无效选项，请重新选择")
```
