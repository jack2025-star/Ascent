```python
#简易通讯录
print("简易通讯录")

#初始化通讯录
contacts  = {
    "小明":{
        "phone":"13800138001",
        "email":"xiaoming@email.com",
        "address":"成都"
    },
    "小红":{
        "phone":"13800138002",
        "email":"xiaohong@email.com",
        "address":"深圳"  
    }

}
```

```python
# 函数显示联系人
def show_contacts():
    print('-'*50)
    #判断字典长度
    if len(contacts) == 0:
        print("联系人为空")
        return
        # 返回None
    
    # 使用for循环遍历,字典名.items()用于获取字典的键值对
    # name 是键    info 是值
    for name,info in contacts.items():
        print(f"姓名:{name}\n")
        print(f"电话:{info['phone']}  ")
        print(f"邮箱:{info['email']}  ")
        print(f"地址:{info['address']}  ")
    
    print(f"共{len(contacts)}个联系人\n")  
    
# 添加联系人
def add_contact():
    print("添加联系人\n")
    name = input("姓名:")
    # 判断是否存在
    if name in contacts:
        print("联系人已存在")
        return
        
    phone = input("电话:")
    email = input("邮箱:")
    address = input("地址:")
    
    contacts[name] = {
        "phone":phone,
        "email":email,
        "address":address
    }
    print("已添加联系人\n")
```

## （4）成绩评级系统
1. 练习 if/elif/else 条件判断
2. 学会处理多个条件分支

```python
#成绩等级系统
print("成绩评级系统")
print("-" * 50)

#获取成绩
try:
    score = float(input("请输入成绩（0-100）："))
    
except ValueError:
    print("\n错误,请输入有效的数字")
    exit()
```

```python
# 检查输入范围
if score < 0 or score > 100:
    print("\n错误：成绩必须在0到100之间,才是合法区间")
    exit()
```

```python
#成绩评级
# 多个 elif 来处理不同的成绩区间
print()

if score >= 90:
    grade = "A"
    level = "优秀"
    comment = "非常棒！继续保持！"

elif score >= 80:
    grade = "B"
    level = "良好"
    comment = "不错！还可以更好！"

elif score >= 70:
    grade = "C"
    level = "中等"
    comment = "还行，需要努力！"

elif score >= 60:
    grade = "D"
    level = "及格"
    comment = "勉强通过，加油！"

else:
    grade = "F"
    level = "不及格"
    comment = "需要加油！"
```

## （5）猜数字游戏

**学习目标：**
1. 掌握条件判断
2. 学会使用while循环
3. 理解随机数的生成

```python
import random  # 导入随机数模块

print("猜数字游戏")

print()

print("游戏规则：")
print("我已经想好了一个 1 到 100 之间的数字")
print("你有3次机会猜中它")
print()

# 生成随机数
# randint(1, 100) 会生成 1到 100之间的随机整数
secret_number = random.randint(1, 100)

# 初始化游戏状态
guess_count = 0       # 已猜次数
max_attempts = 3     # 最大尝试次数
```

```python
# 游戏主循环 
# while 循环：当条件为真时，重复执行

while guess_count < max_attempts:
    # 计算剩余机会
    remaining = max_attempts - guess_count
    try:
        # 获取用户猜测
        guess = int(input(f"第 {guess_count + 1} 次猜测剩余{remaining} 次："))

        # 更新猜的次数
        guess_count += 1

        # 条件判断 
        # 根据猜测结果给出不同提示
        if guess < secret_number:
            print("太小了！再大一点\n")

        elif guess > secret_number:
            print("太大了！再小一点\n")

        else:
            print("🎉恭喜你，猜中了！答案就是",secret_number)
            print(f"你用掉了 {guess_count} 次机会")
            
            break  # 跳出循环，游戏结束

    except ValueError:
        print("错误：请输入有效的数字！\n")
        # 继续循环，不增加次数

# 检查是否失败 
# 如果循环结束后还没猜中
if guess_count >= max_attempts and guess != secret_number:
    print("😞 很遗憾，机会用完了！")
    print(f"正确答案是：{secret_number}")


# --- 游戏结束 ---
print("game over")
```

## （6）待办事项管理器
1. 掌握列表（List）的增删改查
2. 学会使用字典（Dictionary）存储任务
3. 练习循环和条件判断的综合应用

```python
print("📝 待办事项管理器")
print()

# 初始化任务列表
# 使用字典来存储每个任务的详细信息
# 列表来存储多个任务
todos = []  # 空列表，用来存储所有任务

def show_todos():
    print("📋 当前任务列表")
    print()

    if len(todos) == 0:
        print("暂无任务，去休息一下吧！")
        return

    # enumerate 会给列表添加序号
    for i, todo in enumerate(todos, 1):
        status = "✅" if todo['done'] else "⬜"
        print(f"{i}. {status} {todo['task']}")

    print()


```

```python
def add_todo():
    """添加新任务"""
    print("\n添加新任务")
    task = input("请输入任务名称：")

    if task.strip() == "":
        print("任务名称不能为空！")
        return

    # 创建新任务字典
    new_task = {
        'task': task,
        'done': False  # False表示未完成
    }

    # 使用 append() 添加到列表末尾
    todos.append(new_task)
    print(f"已添加任务：{task}")
```

```python
def remove_todo():
    """删除任务"""
    show_todos()

    if len(todos) == 0:
        return

    try:
        num = int(input("请输入要删除的任务编号："))

        # 检查编号是否有效
        if 1 <= num <= len(todos):
            removed = todos.pop(num - 1)  # pop() 删除并返回元素
            print(f"已删除任务：{removed['task']}")
        else:
            print("无效的任务编号！")

    except ValueError:
        print("请输入有效的数字！")
```

```python
def mark_done():
    """标记任务为已完成"""
    show_todos()

    if len(todos) == 0:
        return

    try:
        num = int(input("请输入要标记为完成的任务编号："))

        if 1 <= num <= len(todos):
            # 修改字典中的 done 字段
            todos[num - 1]['done'] = True
            print(f"任务已标记为完成：{todos[num-1]['task']}")
        else:
            print("无效的任务编号！")

    except ValueError:
        print("请输入有效的数字！")

```

```python
# 主程序循环
print("\n欢迎使用待办事项管理器！")

while True:
    print("操作菜单：")
    print("1. 查看所有任务")
    print("2. 添加新任务")
    print("3. 删除任务")
    print("4. 标记完成")
    print("5. 退出")


    choice = input("请选择操作(1-5): ").strip()

    if choice == '1':
        show_todos()

    elif choice == '2':
        add_todo()

    elif choice == '3':
        remove_todo()

    elif choice == '4':
        mark_done()

    elif choice == '5':
        print("\n👋 感谢使用，再见！")
        break

    else:
        print("无效选项，请重新选择！")
```
