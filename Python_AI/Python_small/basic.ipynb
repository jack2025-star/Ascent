## （1）简易计算器
1. 运算符的操作

```python
#简易计算机
import math

try:
    num1 = float(input("请输入第一个数字: "))
    num2 = float(input("请输入第二个数字: "))
except ValueError:
    print("请输入有效的数字")
    exit()

print("请输入运算符:")
print("1加法 +")
print("2减法 -")
print("3乘法 *")
print("4除法 /")
print("5平方根 sqrt")
print("6幂运算 x^y")

choice = input("请输入选项1-6: ")

print()


if choice == '1':
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")

elif choice == '2':
    result = num1 - num2
    print(f"{num1} - {num2} = {result}")

elif choice == '3':
    result = num1 * num2
    print(f"{num1} × {num2} = {result}")

elif choice == '4':
    # 除数不能为 0
    if num2 == 0:
        print("错误：除数不能为0")
    else:
        result = num1 / num2
        print(f"{num1} ÷ {num2} = {result}")
elif choice == '5':
    try:
        result1 = math.sqrt(num1)
        result2 = math.sqrt(num2)
        print(f"结果：{num1} = {result1}, {num2} = {result2}")
    except ValueError:
        print("不能对负数求实数平方根")

elif choice == '6':
     result = num1 ** num2
     print(f"{num1} ^ {num2} = {result}")

else:
    print("错误，请重新输入")
```

## （2）简易通讯录
1. 目标
- 练习字典的使用，字典嵌套字典，数据的查询和修改

2. 实现
- 存储练习人信息 {姓名:{电话,邮箱,地址}}

```python
#简易通讯录
print("简易通讯录")


#初始化通讯录
contacts  = {
    "小明":{
        "phone":"13800138001",
        "email":"xiaoming@email.com",
        "address":"成都"
    },
    "小红":{
        "phone":"13800138002",
        "email":"xiaohong@email.com",
        "address":"深圳"  
    }

}
```

```python
# 函数显示联系人
def show_contacts():
    print('-'*50)
    #判断字典长度
    if len(contacts) == 0:
        print("联系人为空")
        return
        # 返回None
    
    # 使用for循环遍历,字典名.items()用于获取字典的键值对
    # name 是键    info 是值
    for name,info in contacts.items():
        print(f"姓名:{name}\n")
        print(f"电话:{info['phone']}  ")
        print(f"邮箱:{info['email']}  ")
        print(f"地址:{info['address']}  ")
    
    print(f"共{len(contacts)}个联系人\n")  
    
# 添加联系人
def add_contact():
    print("添加联系人\n")
    name = input("姓名:")
    # 判断是否存在
    if name in contacts:
        print("联系人已存在")
        return
        
    phone = input("电话:")
    email = input("邮箱:")
    address = input("地址:")
    
    contacts[name] = {
        "phone":phone,
        "email":email,
        "address":address
    }
    print("已添加联系人\n")
```

```python
#查找联系人
def search_contact():
    print("查询联系人\n")
    name = input("请输入需要查询的联系人：")
    
    if name in contacts:
        info = contacts[name]
        print(f"找到联系人：{name}\n")
        print(f"电话：{info['phone']}  ")
        print(f"邮箱：{info['email']}  ")
        print(f"地址：{info['address']}  ")
    else:
        print(f"未找到联系人：{name}\n")

# 删除联系人
def delete_contact():

    print("删除联系人\n")
    name = input("请输入要删除的姓名：")

    if name in contacts:
        del contacts[name]  # del删除字典中的键
        print(f"已删除联系人：{name}")
    else:
        print(f"未找到联系人：{name}")

#更新联系人
def update_contact():
    print("修改联系人\n")
    name = input("请输入要修改的姓名：")

    if name in contacts:
        print(f"当前信息：\n")
        print(f"电话：{contacts[name]['phone']} ")
        print(f"邮箱：{contacts[name]['email']} ")
        print(f"地址：{contacts[name]['address']} ")

        print("\n请输入新信息,直接回车表示保持原值：")

        phone = input("新电话：")
        if phone:
            contacts[name]['phone'] = phone

        email = input("新邮箱：")
        if email:
            contacts[name]['email'] = aemail

        address = input("新地址：")
        if address:
            contacts[name]['address'] = address

        print(f"已更新联系人：{name}")
    else:
        print(f"未找到联系人：{name}")
```

## （3）成绩评级系统
1. 练习 if/elif/else 条件判断
2. 学会处理多个条件分支

```python
#成绩等级系统
print("成绩评级系统")
print("-" * 50)

#获取成绩
try:
    score = float(input("请输入成绩（0-100）："))
    
except ValueError:
    print("\n错误,请输入有效的数字")
    exit()
```

```python
# 检查输入范围
if score < 0 or score > 100:
    print("\n错误：成绩必须在0到100之间,才是合法区间")
    exit()
```

```python
#成绩评级
# 多个 elif 来处理不同的成绩区间
print()

if score >= 90:
    grade = "A"
    level = "优秀"
    comment = "非常棒！继续保持！"

elif score >= 80:
    grade = "B"
    level = "良好"
    comment = "不错！还可以更好！"

elif score >= 70:
    grade = "C"
    level = "中等"
    comment = "还行，需要努力！"

elif score >= 60:
    grade = "D"
    level = "及格"
    comment = "勉强通过，加油！"

else:
    grade = "F"
    level = "不及格"
    comment = "需要加油！"
```

```python
# 显示结果
print("成绩报告")
print()

print(f"成绩：{score}")
print(f"等级：{grade}")
print(f"评级：{level}")
print(f"评语：{comment}")
```

```python
# 批量处理
# 如果你想同时查看多个成绩的评级
print("\n是否批量查看成绩？(y/n): ")
batch_choice = input().lower() #统一变成小写

if batch_choice == 'y':
    print("\n请输入多个成绩（用空格分隔）：")
    scores_input = input().split()  # split() 按空格分割字符串

    print("批量成绩评级")
    print()

    for s in scores_input:
        try:
            score_val = float(s)
            if 0 <= score_val <= 100:
                if score_val >= 90:
                    g = "A"
                elif score_val >= 80:
                    g = "B"
                elif score_val >= 70:
                    g = "C"
                elif score_val >= 60:
                    g = "D"
                else:
                    g = "F"
                print(f"{score_val:.1f} 对应的等级：{g}") #保留1位小数
            else:
                print(f"{s}无效成绩")
                
        except ValueError:
            print(f"{s}无效输入")
```
