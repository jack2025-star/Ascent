```python
import math

try:
    num1 = float(input("请输入第一个数字: "))
    num2 = float(input("请输入第二个数字: "))
except ValueError:
    print("请输入有效的数字")
    exit()

print("请输入运算符:")
print("1加法 +")
print("2减法 -")
print("3乘法 *")
print("4除法 /")
print("5平方根 sqrt")
print("6幂运算 x^y")

choice = input("请输入选项1-6: ")

print()


if choice == '1':
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")

elif choice == '2':
    result = num1 - num2
    print(f"{num1} - {num2} = {result}")

elif choice == '3':
    result = num1 * num2
    print(f"{num1} × {num2} = {result}")

elif choice == '4':
    # 除数不能为 0
    if num2 == 0:
        print("错误：除数不能为0")
    else:
        result = num1 / num2
        print(f"{num1} ÷ {num2} = {result}")
elif choice == '5':
    try:
        result1 = math.sqrt(num1)
        result2 = math.sqrt(num2)
        print(f"结果：{num1} = {result1}, {num2} = {result2}")
    except ValueError:
        print("不能对负数求实数平方根")

elif choice == '6':
     result = num1 ** num2
     print(f"{num1} ^ {num2} = {result}")

else:
    print("错误，请重新输入")
```
