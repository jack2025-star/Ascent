
## 项目1:成绩统计分析

学习目标：
1. 掌握 NumPy 数组的创建
2. 学会基本的统计函数
3. 理解为什么使用 NumPy


```python
import numpy as np

print("NumPy成绩统计分析")

# 创建数据
# 假设这是一个班级的数学考试成绩
scores = [85, 92, 78, 90, 88, 76, 95, 82, 89, 91]

print("\n原始成绩列表:")
print(scores)
```

```python
# 转换为 NumPy 数组
# NumPy 数组比 Python 列表更高效
scores_array = np.array(scores)

print(f"\n转换为 NumPy 数组后类型：{type(scores_array)}")

# 基本统计
print("\n统计分析")
print(f"最高分：{np.max(scores_array)}")      # 最大值
print(f"最低分：{np.min(scores_array)}")      # 最小值
print(f"平均分：{np.mean(scores_array):.2f}") # 平均值
print(f"中位数：{np.median(scores_array)}")   # 中位数
print(f"总分：{np.sum(scores_array)}")        # 总和
print(f"标准差：{np.std(scores_array):.2f}")  # 标准差（分数离散程度）
```

```python
# 条件筛选
print("\n成绩筛选")

# 找出不及格的成绩（低于 60 分）
failures = scores_array[scores_array < 60]
print(f"不及格成绩：{failures}")

# 找出优秀的成绩（90 分及以上）
excellents = scores_array[scores_array >= 90]
print(f"优秀成绩 (≥90): {excellents}")

# 找出及格的成绩（60-89 分）
passes = scores_array[(scores_array >= 60) & (scores_array < 90)]
print(f"及格成绩 (60-89): {passes}")
```

```python
# 成绩等级划分
def get_grade(score):
    """根据分数返回等级"""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

print("\n学生成绩等级")
for i, score in enumerate(scores_array, 1):
    grade = get_grade(score)
    print(f"学生{i}: {score}分 → 等级{grade}")
```

```python
# 可视化准备
# 计算各个分数段的人数
ranges = [(0, 59), (60, 69), (70, 79), (80, 89), (90, 100)]
print("\n分数段统计")
for low, high in ranges:
    count = len(scores_array[(scores_array >= low) & (scores_array <= high)])
    print(f"{low}-{high}分：{count}人")
print("分析完成！")
```

```python
# ============================================
# 阶段 2 项目 5.2：从 CSV 读取并分析数据
# ============================================
# 学习目标：
#   1. 学会读取 CSV/Excel 文件
#   2. 数据清洗（处理缺失值）
#   3. 更复杂的数据分析
# ============================================

import pandas as pd
import numpy as np

print("=" * 60)
print("📊 CSV 数据读取与分析")
print("=" * 60)

# --- 步骤 1：准备测试数据 ---
# 在实际项目中，你会有一个真实的 CSV 文件

# 这里我们创建一个示例 DataFrame，然后保存为 CSV
data = {
    '姓名': ['张三', '李四', '王五', '赵六', '钱七'],
    '年龄': [25, 30, 28, 35, 22],
    '部门': ['技术部', '销售部', '技术部', '人事部', '销售部'],
    '工资': [15000, 12000, 14000, 10000, 11000],
    '入职日期': ['2020-01-15', '2019-06-20', '2021-03-10', '2018-05-01', '2022-07-15']
}

df = pd.DataFrame(data)

# 保存到 CSV 文件
csv_file = 'employees.csv'
df.to_csv(csv_file, index=False, encoding='utf-8-sig')
print(f"✅ 已创建测试文件：{csv_file}")

# --- 步骤 2：读取 CSV 文件 ---
print("\n--- 读取 CSV 文件 ---")
df_read = pd.read_csv(csv_file)
print(df_read)

# --- 步骤 3：查看数据信息 ---
print("\n--- 数据概览 ---")
print(f"总行数：{len(df_read)}")
print(f"总列数：{len(df_read.columns)}")
print(f"\n数据类型:")
print(df_read.dtypes)

# --- 步骤 4：基本统计 ---
print("\n--- 数值列统计 ---")
print(df_read.describe())

# --- 步骤 5：按部门分析 ---
print("\n--- 按部门统计 ---")
dept_stats = df_read.groupby('部门').agg({
    '工资': ['mean', 'sum', 'count'],
    '年龄': 'mean'
})
print(dept_stats)

# --- 步骤 6：排序 ---
print("\n--- 按工资排序（从高到低）---")
sorted_df = df_read.sort_values('工资', ascending=False)
print(sorted_df[['姓名', '部门', '工资']])

# --- 步骤 7：筛选数据 ---
print("\n--- 筛选：工资大于 13000 的员工 ---")
high_salary = df_read[df_read['工资'] > 13000]
print(high_salary)

print("\n--- 筛选：技术部员工 ---")
tech_dept = df_read[df_read['部门'] == '技术部']
print(tech_dept)

# --- 步骤 8：添加新列 ---
print("\n--- 添加奖金列（工资的 10%）---")
df_read['奖金'] = df_read['工资'] * 0.1
print(df_read[['姓名', '工资', '奖金']])

# --- 步骤 9：计算总收入 ---
total_salary = df_read['工资'].sum()
total_bonus = df_read['奖金'].sum()
print(f"\n--- 财务统计 ---")
print(f"工资总额：¥{total_salary:,.2f}")
print(f"奖金总额：¥{total_bonus:,.2f}")
print(f"支出总计：¥{total_salary + total_bonus:,.2f}")

# --- 步骤 10：保存结果 ---
output_file = 'employees_with_bonus.csv'
df_read.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"\n✅ 结果已保存到：{output_file}")

print("\n" + "=" * 60)
print("数据分析完成！")
print("=" * 60)

# ============================================
# 扩展练习：
# 1. 处理缺失数据（填补缺失值或删除空行）
# 2. 合并多个 CSV 文件
# 3. 生成可视化图表（柱状图、饼图等）
# ============================================
```

```python
# ============================================
# 阶段 2 项目 5.1：消费数据分析报告
# ============================================
# 学习目标：
#   1. 掌握 DataFrame 的创建和查看
#   2. 学会读取 CSV 文件
#   3. 进行基本的数据筛选和统计
# ============================================

import pandas as pd

print("=" * 60)
print("📊 Pandas 消费数据分析")
print("=" * 60)

# --- 步骤 1：创建示例数据 ---
# 在真实项目中，我们会从 CSV 或 Excel 文件读取数据

data = {
    '日期': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05',
            '2024-01-06', '2024-01-07', '2024-01-08', '2024-01-09', '2024-01-10'],
    '类别': ['吃饭', '交通', '购物', '吃饭', '娱乐', '交通', '吃饭', '购物', '吃饭', '娱乐'],
    '金额': [50, 15, 200, 45, 100, 12, 60, 150, 35, 80],
    '支付方式': ['微信', '公交卡', '支付宝', '现金', '微信', '公交卡', '微信', '信用卡', '支付宝', '微信']
}

# 创建 DataFrame（表格）
df = pd.DataFrame(data)

print("\n--- 原始数据 ---")
print(df)

# --- 步骤 2：查看数据基本信息 ---
print("\n--- 数据基本信息 ---")
print(f"行数：{len(df)} 条记录")
print(f"列数：{df.shape[1]} 列")
print(f"列名：{list(df.columns)}")

# --- 步骤 3：总消费统计 ---
total_expense = df['金额'].sum()
avg_expense = df['金额'].mean()
max_expense = df['金额'].max()
min_expense = df['金额'].min()

print("\n--- 总体统计 ---")
print(f"总支出：¥{total_expense:.2f}")
print(f"平均支出：¥{avg_expense:.2f}")
print(f"最大单笔：¥{max_expense:.2f}")
print(f"最小单笔：¥{min_expense:.2f}")

# --- 步骤 4：按类别统计 ---
print("\n--- 按类别统计 ---")
category_stats = df.groupby('类别')['金额'].agg(['sum', 'mean', 'count'])
print(category_stats)

# --- 步骤 5：找出最高消费 ---
max_row = df.loc[df['金额'].idxmax()]
print(f"\n--- 最高消费 ---")
print(f"日期：{max_row['日期']}")
print(f"类别：{max_row['类别']}")
print(f"金额：¥{max_row['金额']:.2f}")
print(f"支付方式：{max_row['支付方式']}")

# --- 步骤 6：筛选特定数据 ---
print("\n--- 筛选：吃饭类消费 ---")
food_expenses = df[df['类别'] == '吃饭']
print(food_expenses)
print(f"吃饭总支出：¥{food_expenses['金额'].sum():.2f}")

print("\n--- 筛选：超过 100 元的消费 ---")
high_expenses = df[df['金额'] > 100]
print(high_expenses)

# --- 步骤 7：按支付方式统计 ---
print("\n--- 按支付方式统计 ---")
payment_stats = df.groupby('支付方式')['金额'].sum()
for payment, amount in payment_stats.items():
    percentage = (amount / total_expense) * 100
    print(f"{payment}: ¥{amount:.2f} ({percentage:.1f}%)")

# --- 步骤 8：保存分析结果 ---
output_file = "expense_analysis.csv"
df.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"\n✅ 数据已保存到：{output_file}")

print("\n" + "=" * 60)
print("分析完成！")
print("=" * 60)

# ============================================
# 扩展练习：
# 1. 添加日期列的月份提取
# 2. 计算每个类别的占比饼图
# 3. 导入真实的 CSV 数据进行分析
# ============================================
```

```python
# ============================================
# 阶段 2 项目 6.1：数据可视化分析
# ============================================
# 学习目标：
#   1. 学习使用 matplotlib 绘图
#   2. 创建柱状图、饼图、折线图
#   3. 生成专业的数据报告
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

print("=" * 60)
print("📈 数据可视化分析")
print("=" * 60)

# --- 步骤 1：准备数据 ---
sales_data = {
    '月份': ['1 月', '2 月', '3 月', '4 月', '5 月', '6 月'],
    '销售额': [120, 150, 180, 200, 220, 250],
    '成本': [80, 90, 100, 110, 120, 130]
}

df = pd.DataFrame(sales_data)
print("\n销售数据:")
print(df)

# --- 步骤 2：计算利润 ---
df['利润'] = df['销售额'] - df['成本']
print("\n加上利润列后:")
print(df)

# --- 步骤 3：创建柱状图 ---
plt.figure(figsize=(10, 6))
plt.bar(df['月份'], df['销售额'], color='skyblue', label='销售额')
plt.bar(df['月份'], df['成本'], color='salmon', label='成本')
plt.xlabel('月份')
plt.ylabel('金额（万元）')
plt.title('半年度销售 vs 成本对比')
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.savefig('sales_bar_chart.png', dpi=150, bbox_inches='tight')
print("\n✅ 已保存柱状图：sales_bar_chart.png")
plt.close()

# --- 步骤 4：创建折线图 ---
plt.figure(figsize=(10, 6))
plt.plot(df['月份'], df['销售额'], marker='o', color='blue', linewidth=2, label='销售额')
plt.plot(df['月份'], df['利润'], marker='s', color='green', linewidth=2, label='利润')
plt.xlabel('月份')
plt.ylabel('金额（万元）')
plt.title('半年度销售和利润趋势')
plt.legend()
plt.grid(alpha=0.3)

plt.savefig('sales_line_chart.png', dpi=150, bbox_inches='tight')
print("✅ 已保存折线图：sales_line_chart.png")
plt.close()

# --- 步骤 5：创建饼图（分类占比）---
category_data = {
    '电子产品': 35,
    '服装': 25,
    '食品': 20,
    '家居': 15,
    '其他': 5
}

plt.figure(figsize=(8, 8))
categories = list(category_data.keys())
values = list(category_data.values())
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c9c9ff']

plt.pie(values, labels=categories, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('各品类销售占比')

plt.savefig('sales_pie_chart.png', dpi=150, bbox_inches='tight')
print("✅ 已保存饼图：sales_pie_chart.png")
plt.close()

# --- 步骤 6：多子图展示 ---
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# 左上：销售额柱状图
axes[0, 0].bar(df['月份'], df['销售额'], color='steelblue')
axes[0, 0].set_title('月度销售额')
axes[0, 0].set_ylabel('金额（万元）')

# 右上：利润折线图
axes[0, 1].plot(df['月份'], df['利润'], marker='o', color='green', linewidth=2)
axes[0, 1].set_title('月度利润趋势')
axes[0, 1].set_ylabel('金额（万元）')

# 左下：成本占比饼图
total_cost = df['成本'].sum()
cost_percent = (df['成本'] / total_cost * 100).tolist()
axes[1, 0].pie(df['成本'], labels=df['月份'], autopct='%1.1f%%')
axes[1, 0].set_title('月度成本占比')

# 右下：销售增长散点图
growth = df['销售额'].diff().fillna(0)
axes[1, 1].scatter(range(len(growth)), growth, color='orange', s=100)
axes[1, 1].set_title('月度增长情况')
axes[1, 1].set_xlabel('月份序号')
axes[1, 1].set_ylabel('增长额')

plt.tight_layout()
plt.savefig('sales_dashboard.png', dpi=150, bbox_inches='tight')
print("✅ 已保存综合仪表盘：sales_dashboard.png")
plt.close()

# --- 步骤 7：生成分析报告 ---
print("\n" + "=" * 60)
print("📊 销售分析报告")
print("=" * 60)
print(f"\n总销售额：¥{df['销售额'].sum():,.2f} 万")
print(f"总成本：¥{df['成本'].sum():,.2f} 万")
print(f"总利润：¥{df['利润'].sum():,.2f} 万")
print(f"平均利润率：{(df['利润'].sum() / df['销售额'].sum() * 100):.1f}%")
print(f"\n最高销售月份：{df.loc[df['销售额'].idxmax(), '月份']} ({df['销售额'].max()}万)")
print(f"最低销售月份：{df.loc[df['销售额'].idxmin(), '月份']} ({df['销售额'].min()}万)")
print(f"平均月度增长：{df['销售额'].diff().mean():.2f} 万")

print("\n" + "=" * 60)
print("✅ 所有图表已生成！")
print("=" * 60)

# ============================================
# 扩展练习：
# 1. 添加更多图表类型（如面积图、雷达图）
# 2. 交互式图表（使用 plotly）
# 3. 导出 PDF 报告
# ============================================
```
