
## 项目1:成绩统计分析

学习目标：
1. 掌握 NumPy 数组的创建
2. 学会基本的统计函数
3. 理解为什么使用 NumPy


```python
import numpy as np

print("NumPy成绩统计分析")

# 创建数据
# 假设这是一个班级的数学考试成绩
scores = [85, 92, 78, 90, 88, 76, 95, 82, 89, 91]

print("\n原始成绩列表:")
print(scores)
```

```python
# 转换为 NumPy 数组
# NumPy 数组比 Python 列表更高效
scores_array = np.array(scores)

print(f"\n转换为 NumPy 数组后类型：{type(scores_array)}")

# 基本统计
print("\n统计分析")
print(f"最高分：{np.max(scores_array)}")      # 最大值
print(f"最低分：{np.min(scores_array)}")      # 最小值
print(f"平均分：{np.mean(scores_array):.2f}") # 平均值
print(f"中位数：{np.median(scores_array)}")   # 中位数
print(f"总分：{np.sum(scores_array)}")        # 总和
print(f"标准差：{np.std(scores_array):.2f}")  # 标准差（分数离散程度）
```

```python
# 条件筛选
print("\n成绩筛选")

# 找出不及格的成绩（低于 60 分）
failures = scores_array[scores_array < 60]
print(f"不及格成绩：{failures}")

# 找出优秀的成绩（90 分及以上）
excellents = scores_array[scores_array >= 90]
print(f"优秀成绩 (≥90): {excellents}")

# 找出及格的成绩（60-89 分）
passes = scores_array[(scores_array >= 60) & (scores_array < 90)]
print(f"及格成绩 (60-89): {passes}")
```

```python
# 成绩等级划分
def get_grade(score):
    """根据分数返回等级"""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

print("\n学生成绩等级")
for i, score in enumerate(scores_array, 1):
    grade = get_grade(score)
    print(f"学生{i}: {score}分 → 等级{grade}")
```

```python
# 可视化准备
# 计算各个分数段的人数
ranges = [(0, 59), (60, 69), (70, 79), (80, 89), (90, 100)]
print("\n分数段统计")
for low, high in ranges:
    count = len(scores_array[(scores_array >= low) & (scores_array <= high)])
    print(f"{low}-{high}分：{count}人")


print("分析完成！")
```
