```python
import numpy as np
```

（1）ndarray对象的索引和切片与原生Python中的索引和切片相同，索引是指获取数组中某个特定位置的一个元素；切片是指获取数组中元素的子集（部分数组）。  
（2）ndarray对象的索引和切片序号体系也包括从**0开始的正向递增序号**和**从-1开始的反向递减序号**。

# 1. 一维数组的索引和切片

一维数组的索引和切片与原生python的列表完全相同。

## 1.1 一维数组的索引

```python
a = np.arange(21, 31)
a
```

```python
# 分别取出索引为0,3，-1，-3的值
a[0],a[3],a[-1],a[-3]
```

## 1.2 一维数组的切片

```python
a = np.arange(21, 31)
a
```

```python
# 从索引为1取到索引为8，从索引为-8取到索引为-1
a[1:8],a[-8:-1]
```

# 2. 多维数组的索引和切片

多维数组请注意：其维度应该从外层，往里层看，即最外层为第一维度。

## 2.1 多维数组的索引

```python
a = np.arange(30).reshape(2, 3, 5)
a
```

```python
# 取出元素7
```

```python
# 正向反向取出元素28
```

## 2.2 多维数组的切片

```python
a = np.arange(30).reshape(2, 3, 5)
a
```

# 4. 总结

（1）ndarray对象的索引和切片与原生Python中的索引和切片相同，索引是指获取数组中某个特定位置的元素；切片是指获取数组中元素的子集。  
（2）ndarray对象的索引和切片序号体系也包括从0开始的正向递增序号和从-1开始的反向递减序号。  
（3）一维ndarray对象的索引和切片与原生Python中的索引和切片几乎完全一样。  
（4）多维ndarray对象的索引和切片维度应从最外层开始向里层看
