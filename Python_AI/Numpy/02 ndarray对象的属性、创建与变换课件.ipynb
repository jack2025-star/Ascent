# 1. ndarray对象简介

（1）NumPy 最重要的一个特点是其 N 维数组对象 ndarray，它是一系列同类型数据的集合，以 0 下标为开始进行集合中元素的索引。
（2）ndarray 对象是用于存放同类型元素的多维数组。  
（3）ndarray 中的每个元素在内存中都有相同存储大小的区域。    
（4）Numpy的ndarray对象与原生Python的list对象很相似，实现的功能也基本相同，区别在于：  
1）ndarray对象计算性能更高效。  
2）ndarray对象一般要求所有元素是**同一种数据类型**，如都是数字类型，或都是字符串，或其他，否则可能会报错；而list对象的元素可以是任何数据类型。

```python
li = [1,2,'3',4,'a']
li
```

```python
np_array = np.array([1,2.5,"3"])
np_array
```

# 2. ndarray对象的属性

（1）ndim：ndarray对象的维度。  
（2）shape：ndarray对象的形状，对于矩阵，即行数和列数。   
（3）size：ndarray对象的元素个数，相当于array.shape中每个数值的乘积。  
（4）dtype：ndarray对象的元素的数据类型。  
（5）itemsize：ndarray对象中每个元素的大小，以字节为单位。

```python
a2 = np.arange(1, 21).reshape(4, 5)
a2
```

```python
f'数组a2的维度是：{a2.ndim}；形状是：{a2.shape}；元素个数是：{a2.size}；元素的数据类型是：{a2.dtype}；元素的大小是：{a2.itemsize}。'
```

1. 维度是：2  
对应属性：a2.ndim  
含义：表示数组的轴数。a2 是一个二维数组（有行和列），所以是 2。如果是纯数字（单个值），维度是 0；如果是列表（一排数），维度是 1。  
2. 形状是：(4, 5)  
对应属性：a2.shape  
含义：表示数组在各个维度上的大小。这是一个元组（Tuple）。对于 a2，它表示“第一维（行）有 4 个单位，第二维（列）有 5 个单位”。  
3. 元素个数是：20  
对应属性：a2.size  
含义：数组中所有元素的总数量。它等于形状元组中各个数字的乘积（$4 \times 5 = 20$）。  
4. 元素的数据类型是：int64  
对应属性：a2.dtype  
含义：数组中每个元素的数据类型。因为 a2 存储的是整数，在 64 位系统上 NumPy 通常默认将其创建为 int64（64 位整数）。  
5. 元素的大小是：8  
对应属性：a2.itemsize  
含义：单个元素在内存中占用的字节数（Bytes）。  
计算逻辑：int64 表示 64 位（bit），而 8 位等于 1 字节，所以64÷8=8 字节。如果是 float32，这里通常会显示 4

## 3.2 使用内置函数创建ndarray对象

### 3.2.1 数字范围数组
#### 3.2.1.1使用np.arange()函数创建

（1）np.arange([start],[stop],[step],dtype=None)  
（2）np.arange()函数和原生Python中的range()函数类似，都是左闭右开区间[start,stop)。  
（3）np.arange()函数的step支持浮点数，原生Python的range()函数不支持。  
（4）np.arange()函数**默认只能创建一维数组**，但是可以结合**reshape()函数**将其转换成多维数组。不过，需要保证新形状的乘积与元素总数相同。

```python
# 生成0-9的1维数组
np.arange(0,10)
```

```python
# 生成1-11步长为2
np.arange(1,12,2)
```

```python
# 生成1-11步长为0.5
np.arange(1,12,0.5)
```

```python
# 生成0-9的2维数组
np.arange(0,10).reshape(5,2)
```

```python
# 生成1-18的3维数组
np.arange(1,19).reshape(3,2,3)
```

**数据连续性：reshape 操作是按顺序“切片”填充的。它先把前 6 个数（1-6）填满第一层，再把接下来的 6 个数（7-12）填满第二层，以此类推。  
形状兼容性：你必须确保 arange 生成的元素总数（这里是 18）能够被 reshape 的维度整除。如果你写成 .reshape(3, 3, 3)，程序会报错，因为 3×3×3=27，与 18 不匹配。**

#### 3.2.1.2使用np.linspace()函数创建

(1)np.linspace(start, stop, num=50, endpoint=True, retstep=False,dtype=None, axis=0)  
(2)np.linspace() 返回指定间隔内的等距数字。返回在 [start，stop] 间隔内计算的等距采样数。可以选择排除间隔的端点。

```python
# 生成2-3之间5个等距的数组
np.linspace(2,3,5)
```

```python
# 生成2-3之间不包含3，5个等距的数组
np.linspace(2,3,5,endpoint = False)
```

### 3.2.1 使用np.ones()/np.ones_like()、np.zeros()/np.zeros_like()、np.empty()/np.empty_like()、np.full()/np.full_like()、np.eye()等函数创建

（1）np.ones()、np.zeros()、np.empty()、np.full()等函数是**独立**创建一个新的ndarray对象，对象中的元素分别全为1、0、空及指定数值。  
（2）np.ones_like()、np.zeros_like()、np.empty_like()、np.full_like()等是**在当前某个数组基础上**，参考该数组的形状样式（shape），创建一个新的ndarray对象，对象中的元素分别全为1、0、空及指定数值。    
（3）np.eye()是指创建**单位矩阵（对角线为1，其他值为0）**。 

```python
# 1行10列全为1
np.ones((1,10))
```

```python
# 2行5列全为1
  # 注意：ones的参数必须是一个元组如(2,5)，不能直接写2,5
np.ones((2,5))
```

```python
# 1行5列全为0
np.zeros((1,5))
```

```python
# 3行4列全为0
np.zeros((3,4))
```

```python
# 返回3行6列内容随机
np.random.rand(3,6)
```

```python
# full(shape, fill_value, dtype, order, like)
# 生成1行10列，全为6的数组
np.full((1,10),6)
```

```python
# 生成2行5列全为6的数组
np.full((2,5),6)
```

```python
# 生成2行2列值为[1,2]的数组
np.full
```

```python
# eye(N, M, k, dtype, order, like)，N行M列N，k是指第一行元素中，数字1的索引，默认是0。
# 生成5行5列对角线为1的数组
np.eye(5,5)
```

N=5 (行数)：  
指定生成矩阵的行数为 5。  
M=5 (列数)：  
指定生成矩阵的列数为 5。  
注：通常如果不传入第二个参数（即只写 np.eye(5)），M 会默认等于 N，生成一个方阵。显式写出 5, 5 可以让代码意图更清晰，或者用于生成非方阵（如 np.eye(3, 4)）。

```python
# 生成5行5列移动对角线的数组
np.eye(5,5,k = 1)

# 2. 向右移动1位 (k=1) - 上移对角线
# 3. 向左移动1位 (k=-1) - 下移对角线
```

## 3.3 使用np.random模块下的方法进行创建

| 函数 | 分布类型 | 数值范围 | 主要特点 |
|:-----|:---------|:---------|:---------|
| `np.random.random(size)` | 均匀分布 | `[0.0, 1.0)` | 参数为元组形式的形状，生成0到1之间的随机数 |
| `np.random.uniform(low, high, size)` | 均匀分布 | `[low, high)` | 可以自定义范围的均匀分布随机数 |
| `np.random.randint(low, high, size)` | 均匀分布 | `[low, high)` 的整数 | 生成随机整数 |
| `np.random.rand(d0, d1, ...)` | 均匀分布 | `[0.0, 1.0)` | 参数直接定义形状，生成0到1之间的随机数 |
| `np.random.randn(d0, d1, ...)` | 标准正态分布 | 任意实数（均值为0，标准差为1） | 参数直接定义形状，生成符合标准正态分布的随机数 |

## 4.2 ndarray对象形状、维度等变换

| 方法 | 说明 |
|:---|:---|
| np.reshape(shape[, order]) | 返回包含具有新形状的相同数据的数组|
| np.resize(new_shape[, refcheck]) | 更改阵列的形状和大小，重新分配内存|
| np.transpose(*axes) | 返回轴已转置的数组视图(view) |
| np.swapaxes(axis1, axis2) | 返回轴1与轴2水平互换后的视图(view)|
| np.flatten([order]) | 返回折叠为一维的数组的副本 |
| np.ravel([order]) | 返回展平数组 |
