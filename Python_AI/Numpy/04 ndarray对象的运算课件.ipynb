# 1. ndarray对象的运算

注意：此处所讲的运算主要是基本的数组运算（即对应**位置**上的数据进行运算），不考虑矩阵的乘法运算（点积运算）。如果想要实现矩阵的乘法运算，可将ndarray对象转换成matrix对象，然后再使用\*号进行运算；也可以直接使用np.dot()方法对两ndarray对象进行矩阵乘法。如np.matrix(a1)*np.matrix(a2)；np.dot(a1,a2)

## 1.1 广播（Broadcast）运算机制

ndarray对象支持广播（Broadcast）运算机制。  
（1）当参与运算的两个ndarray对象形状（shape）**相同**时，其运算逻辑就是**对应位置**上的元素进行运算。  
（2）当参与运算的两个ndarray对象形状（shape）**不相同，且满足广播机制的条件**时，则会触发广播机制。  
（3）当参与运算的两个ndarray对象形状（shape）**不相同，且不满足广播机制的条件**，则不会触发广播机制，并会报错ValueError（即无法正常运算）。  
**（4）广播机制的条件：**
+ 两个数组各维度大小**从后往前**比对，低维度数组的各维度大小与高维度数组的各维度大小均相同（注意：广播是从内向外的扩展元素的，因此需要将两个元素从后往前对比）。如(2,3,4,5)和(3,4,5)、(4,5)。  
+ 如果两个数组各维度大小从后往前对比时，存在某些维度的大小不相等时，但其中一个数组**该不相等维度大小为1**。如(2,3,4,5)和(2,3,1,5)、(3,1,5)。

```python
a = np.arange(1, 9).reshape(2, 4)  # (2,4)
b = 10 # 标量
c = np.array([10, 11, 12, 13])  # (1,4)
d = np.arange(21, 29).reshape(2, 4)  # (2,4)
e = np.arange(31, 43).reshape(3, 4)  # (3,4)
f = np.arange(41, 49).reshape(4, 2)  # (4,2)
g = np.arange(51, 57).reshape(2, 3)  # (2,3)
```

# 2. ndarray对象中轴（axis）的概念

（1）axis与维度的概念类似，ndim表示轴的数量或维度的数量。一维数组有一个轴，即axis=0，以此类推，二维数据有两个轴，分别是axis=0，axis=1。
