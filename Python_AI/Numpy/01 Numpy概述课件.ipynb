# 1. Numpy简介

（1）Numpy（Numerical Python）是Python的一个科学计算库，支持**多维数组和矩阵运算**，同时也提供了很多**数学函数**，包括线性代数、傅里叶变换、随机数生成等。  
（2）Numpy以**数组或矩阵为“粒度”**进行计算，并且支持大量的数学函数，因此，与原生Python（for循环）相比，使用Numpy的**代码更简洁、性能更高效**。    
（3）Numpy是很多Python科学计算库的**基础库**，如Scipy、pandas、Scikit-Learn、Tensorflow等。

## 3.2 两个数组求和

a4是数字0到n-1中每一个数的**平方**组成的数组；a5是数字0到n-1中每一个数的**立方**组成的数组；计算数组a4与数组a5的和。

### 3.2.1 使用原生Python代码实现数组加法

```python
# 列表推导式
# n = 5
# result = [i**2 + i**3 for i in range(n)]
# print(result)

def array_python(n):
    a1 = [i** 2 for i in range(n)]
    a2 = [i** 3 for i in range(n)]
    a3 = [a+ b for a,b in zip(a1,a2)]
    return a3
```

### 3.2.2 使用Numpy库实现数组加法

```python
# i = np.arange(5)
# a4 = i ** 2
# a5 = i ** 3
# result1 = a4 + a5
# print(result1)

def array_numpy(n):
    a1 = np.arange(n)** 2
    a2 = np.arange(n)** 3
    a3 = a1+ a2
    return a3
```

### 3.3.3 测试性能

分别使n=100（百），n=10000（万），n=10000000（千万）次，对比执行效率。

```python
# import time
# n = 10000000
# time_py = time.time()
# result_list = [i ** 2 + i **3 for i in range(n)]
# list_time = time.time() - time_py
# print(list_time)

%timeit array_python(10000)
%timeit array_numpy(10000)
```
